### Sorting Search Terms By Priority

```python
order = ["HIGH", "MEDIUM", "LOW"]
search_terms["Priority"] = pd.Categorical(search_terms["Priority"], categories=order, ordered=True)
```

```python
search_terms = search_terms.sort_values("Priority", ignore_index=True)
search_terms
```
