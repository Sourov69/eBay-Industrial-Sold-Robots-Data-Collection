## 9. Creating Column : **brand_name**

```python
df["listing_title"]
```

```python
brand_list = ["FANUC", "ABB", "KUKA", "Yaskawa", "Universal Robots", "Kawasaki", "Epson", "Denso", "Mitsubishi", "Staubli", "Comau", "Nachi"]
brand_list
```

```python
df["listing_title"][0]
```

```python
def search_brand(listing_title):
    brand_list = ["FANUC", "ABB", "KUKA", "Yaskawa", "Universal Robots",
                  "Kawasaki", "Epson", "Denso", "Mitsubishi",
                  "Staubli", "Comau", "Nachi"]

    if pd.isna(listing_title):  # handle missing values
        return None

    title = listing_title.upper()

    for brand in brand_list:
        if brand.upper() in title:
            return brand.upper()

    return None
```

```python
df["listing_title"].apply(search_brand)
```

```python
df["brand_name"] = df["listing_title"].apply(search_brand)
```
